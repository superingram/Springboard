# **Job search word cloud**

This short notebook produces a word cloud in Python from a text file of several job descriptions that has the capability to show phrases depending on the collocation threshold set.  In other words, if the collocation threshold is set higher, then the word cloud will mostly display single words.  If the threshold is set to numbers less than 10, the word cloud will display mostly phrases. We will do both.  Additionally, a complete list of important word combinations will be provided.

### **Important packages**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from textblob import TextBlob
```

### **Reading in the text file of job descriptions**

```python
df = open("/Users/brian/Desktop/Wordcloud.txt", "r", encoding="ISO-8859-1")

print(df.read(20))

f = df.read()
```

### **Text preparation**

```python
# Converting the sample text to a blob
SampleTextInBlobFormat = TextBlob(f)
 
# Finding the noun phrases (important keywords combination) in the text
# This can help to find out what entities are being talked about in the given text
NounPhrases=SampleTextInBlobFormat.noun_phrases
 
# Creating an empty list to hold new values
# combining the noun phrases using underscore to visualize it as wordcloud
NewNounList=[]
for words in NounPhrases:
    NewNounList.append(words.replace(" ", "_"))
    
# Converting list into a string to plot wordcloud
NewNounString=' '.join(NewNounList)
```

### **Word cloud construction - single words**

```python
stopwords = set(STOPWORDS)

stopwords.update(["data", "scientist", "upstart", "computer", "statistics", "data_scientist", "data_science", "u"])

wordcloud = WordCloud(max_words=70, background_color ='white', 
            stopwords = stopwords, collocation_threshold = 15,  
            min_font_size = 10).generate(NewNounString)

plt.figure(figsize = (12, 12), facecolor = None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad = 0) 

plt.show()
```

### **Word cloud construction - phrases**

```python
wordcloud = WordCloud(max_words=70, background_color ='white', 
            stopwords = stopwords, collocation_threshold = 4,  
            min_font_size = 12).generate(NewNounString)

plt.figure(figsize = (12, 12), facecolor = None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad = 0) 

plt.show()
```
